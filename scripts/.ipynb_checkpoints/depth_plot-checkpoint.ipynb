```python
import numpy as np
import matplotlib.pyplot as plt

path =  "/home/xihelm/catkin_ws/src/tomato_grasp/compare_data/"

depth = np.load(path + "d435_depth.npy", allow_pickle=True)
%matplotlib inline

np.random.seed(42)

depth
```

```python
plt.hist(depth, ec="purple", fc="green", alpha=0.5)
plt.xlabel('distance (mm)')
plt.ylabel('frequency')
plt.title('Full Tray Depth Histogram for realsense D435')
plt.show()
```

```python
depth = np.load(path + "d405_depth.npy", allow_pickle=True)
%matplotlib inline

np.random.seed(42)

plt.hist(depth, ec="purple", fc="green", alpha=0.5)
plt.xlabel('distance (mm)')
plt.ylabel('frequency')
plt.title('Full Tray Depth Histogram for realsense D405')
plt.show()
```

```python
depth = np.load(path + "zed2i_depth.npy", allow_pickle=True)
%matplotlib inline

np.random.seed(42)
depth =np.nan_to_num(depth, nan=0, posinf=-1)
depth[depth > 100] = -1
depth[depth!=0]
plt.hist(depth, ec="purple", fc="green", alpha=0.5)
plt.xlabel('distance (m)')
plt.ylabel('frequency')
plt.title('Full Tray Depth Histogram for zed2i')
plt.show()
```

```python
depth = np.load(path + "zedm_depth.npy", allow_pickle=True)
%matplotlib inline

np.random.seed(42)
depth =np.nan_to_num(depth, nan=0, posinf=-1, neginf=-1)
depth[depth > 100] = -1
depth[depth!=0]
plt.hist(depth, ec="purple", fc="green", alpha=0.5)
plt.xlabel('distance (m)')
plt.ylabel('frequency')
plt.title('Full Tray Depth Histogram for zedm')
plt.show()
```
