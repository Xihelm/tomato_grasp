```python
import numpy as np
import matplotlib.pyplot as plt

path =  "/home/xihelm/catkin_ws/src/tomato_grasp/compare_data/"

depth = np.load(path + "d435_depth.npy", allow_pickle=True)
%matplotlib inline

np.random.seed(42)

depth
```
